```python
img = czifile.imread('../proprietary_files/Osteosarcoma_01.czi')

print(img.shape)  #7 dimensions
#Time series, scenes, channels, x, y, z, RGB
#IN this example (Osteosarcoma) we have 1 time series, 1 scene, 3 channels and each channel grey image
#size 1104 x 1376

#Let us extract only relevant pixels, all channels in x and y
img1=img[0, 0, :, :, :, 0]
print(img1.shape)

#Next, let us extract each channel image.
img2=img1[0,:,:]  #First channel
img3=img1[1,:,:] #Second channel
img4=img1[2,:,:] #Third channel
#io.imshow(img4)
```

```python
plt.imshow(img2, cmap='hot')
```

```python
plt.imshow(img3, cmap='hot')
```

```python
plt.imshow(img4, cmap='hot')
```

```python
import numpy as np

# Nomalize image's pixel value
img2 = np.clip(img2 / (img2.max() / 255), 0, 255).astype(np.uint8)
img3 = np.clip(img3 / (img3.max() / 255), 0, 255).astype(np.uint8)
img4 = np.clip(img4 / (img4.max() / 255), 0, 255).astype(np.uint8)

rgb_image = np.stack((img2, img3, img4), axis=-1)
plt.imshow(rgb_image)
```
